```python
from datetime import datetime   
from dateutil.relativedelta import relativedelta

import pandas as pd
pd.set_option("display.max_columns", 30) 
import requests
```

```python
current_datetime = datetime.now() - relativedelta(months=2)
formatted_datetime = current_datetime.strftime("%Y-%m-%d")

url = ( 
    f"https://data.cityofchicago.org/resource/ajtu-isnz.json?"
    f"$where=trip_start_timestamp >= '{formatted_datetime}T00:00:00' " 
    f"AND trip_start_timestamp <= '{formatted_datetime}T23:59:59'"
    f"&$limit=30000"
)
response = requests.get(url)
data = response.json()

len(data)  # Number of taxi trips on February 16, 2024
```

```python
taxi_trips = pd.DataFrame(data)
taxi_trips.head()
```

#### Transformation : deal with NaN values

```python
taxi_trips.drop(["pickup_census_tract", "dropoff_census_tract"], axis=1, inplace=True)
```

```python
taxi_trips.drop(["pickup_centroid_location", "dropoff_centroid_location"], axis=1, inplace=True)
```

```python
taxi_trips.dropna(inplace=True)
```

```python
taxi_trips.info()
```

#### Transformation: renaming

```python
taxi_trips.rename(columns={"pickup_community_area": "pickup_area_id",
                            "dropoff_community_area": "dropoff_area_id"}, inplace=True)
```

```python
taxi_trips.columns
```

#### Transformation: create helper column for weather data

```python
taxi_trips['trip_start_timestamp'] = pd.to_datetime(taxi_trips['trip_start_timestamp'])
```

```python
taxi_trips["datetime_for_weather"] = taxi_trips["trip_start_timestamp"].dt.floor("h")
```

```python
taxi_trips["datetime_for_weather"].unique()
```

#### Check: join taxi and weather data

```python
current_datetime = datetime.now() - relativedelta(months=2)
formatted_datetime = current_datetime.strftime("%Y-%m-%d")

formatted_datetime



url = "https://archive-api.open-meteo.com/v1/era5"

params = {
    "latitude": 41.85,
    "longitude": -87.65,
    "start_date": formatted_datetime,
    "end_date": formatted_datetime,
    "hourly":"temperature_2m,wind_speed_10m,rain,precipitation",
}

response = requests.get(url, params=params)
data = response.json()

weather_data = {
    "datetime":data["hourly"]["time"],
    "temperature":data["hourly"]["temperature_2m"],
    "wind_speed":data["hourly"]["wind_speed_10m"],
    "rain": data["hourly"]["rain"],
    "precipitation":data["hourly"]["precipitation"],
    }


weather_df = pd.DataFrame(weather_data)
weather_df["datetime"] = pd.to_datetime(weather_df["datetime"])

weather_df.head()
```

```python
taxi_trips_with_weather = taxi_trips.merge(weather_df, 
                                          left_on="datetime_for_weather", right_on="datetime")
```

```python
taxi_trips_with_weather.head()
```

#### Transformation: data type conversions

```python
taxi_trips.info()
```

```python
taxi_trips_original = taxi_trips.copy()
```

```python
data_types = {
    "trip_end_timestamp": "datetime64[ns]",
    "trip_seconds": "int32",
    "trip_miles": "float",
    "pickup_area_id": "int8",
    "dropoff_area_id": "int8",
    "fare": "float",
    "tips": "float",
    "tolls": "float",
    "extras": "float",
    "trip_total": "float",
    "pickup_centroid_latitude": "float",
    "pickup_centroid_longitude": "float",
    "dropoff_centroid_latitude": "float",
    "dropoff_centroid_longitude": "float",
}

taxi_trips = taxi_trips.astype(data_types)
```

##### Memory usage

```python
orig_mem = taxi_trips_original.memory_usage(deep=True).sum() 
opt_mem = taxi_trips.memory_usage(deep=True).sum()

reduction = (orig_mem - opt_mem) / orig_mem * 100

print(f"Original: {orig_mem:,} bytes")
print(f"Optimized: {opt_mem:,} bytes")
print(f"Reduction: {reduction:.2f}%")
```
